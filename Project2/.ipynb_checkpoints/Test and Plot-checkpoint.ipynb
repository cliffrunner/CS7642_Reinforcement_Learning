```python
import json
import os
import matplotlib.pyplot as plt
folder = 'outputs'
with open(os.path.join(folder, 'rewards.json'), 'r') as handle:
    rewards = json.load(handle)
```

```python
plt.figure(figsize=(8,5))
plt.plot(list(rewards.values())[0])
plt.xlabel('episode')
plt.ylabel('total reward')
```

```python
with open(os.path.join(folder, 'rewards100.json'), 'r') as handle:
    rewards = json.load(handle)
plt.figure(figsize=(8,5))
plt.plot(list(rewards.values())[0])
plt.xlabel('episode')
plt.ylabel('total reward')
```

```python
plt.hist(list(rewards.values())[0], bins=10, rwidth=0.9, label='histogram')
plt.hist(list(rewards.values())[0], bins=100, cumulative=True, histtype='step', label='cumulative distribution')
plt.xlabel('total reward')
plt.ylabel('occurrence')
plt.legend(loc='upper left')
```
