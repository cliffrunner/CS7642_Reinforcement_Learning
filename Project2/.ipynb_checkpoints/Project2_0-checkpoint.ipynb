```python
import numpy as np
import gym
import os
import math
import json
import copy
import itertools
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam

from tqdm import tqdm
from sklearn.preprocessing import StandardScaler
from collections import deque
env = gym.make('LunarLander-v2')

gamma = 0.99
alpha = 0.1
lr = 0.0005

epsilon = 1
d_epsilon = 0.998
n_episodes = 2000
max_step = 1500
history_len = int(1e5)
n_states = env.observation_space.shape[0]
n_actions = env.action_space.n

optimizer = Adam(lr=lr)

folder = 'outputs_' + \
        'alpha{}_'.format(str(alpha)) + \
        'lr{}_'.format(str(lr)) + \
        'gamma{}'.format(str(gamma))
os.makedirs(folder, exist_ok=True)
```

```python
class DLModel(object):
    def initialize_model(self, 
                         activation='relu', 
                         n_inputs=8, 
                         n_outputs=4, 
                         loss='mean_squared_error',
                         optimizer=optimizer):
        model = Sequential()
        model.add(Dense(64, input_dim=n_inputs, activation=activation))
        model.add(Dense(64, activation=activation))
        model.add(Dense(n_outputs))
        model.compile(loss=loss, optimizer=optimizer)
        return model
```

```python
class Agent(DLModel):
    def __init__(self, gamma, epsilon, d_epsilon, alpha, n_actions, n_states, history_len):
        self.gamma = gamma
        self.epsilon = epsilon
        self.d_epsilon = d_epsilon
        self.alpha = alpha
        self.n_actions = n_actions
        self.n_states = n_states
        self.history = deque(maxlen=history_len)
        self.n = 0
        self.model = self.initialize_model(n_inputs=self.n_states, n_outputs=self.n_actions)
    def epsilon_decay(self):
        self.epsilon *= self.d_epsilon
        if self.epsilon <= 0.05: 
            self.epsilon==0.05
        pass
    def select_action(self, s):
        s = s.reshape(1,-1)
        if np.random.random() > self.epsilon:
            action = np.argmax(self.model.predict(s))
        else:
            action = np.random.randint(0, self.n_actions)
        return action
    def append_history(self, state, action, reward, state_next, done):
        self.history.append([state, action, reward, state_next, done])
    def fit(self, train_set_size=20, steps=5):
        self.n += 1
        if self.n%steps!=0: 
            return
        self.n = 0
        if len(self.history)<train_set_size: 
            return
        
        actualQ = np.zeros((train_set_size,self.n_actions))
        Q = np.zeros((train_set_size,self.n_actions))
        idx = np.random.choice(np.array(len(self.history)), size=train_set_size, replace=False)
        history = [self.history[_] for _ in idx]
        States = np.array([_[0] for _ in history])
        actions = []
        idx = []
        for i, stuff in enumerate(history):
            idx.append(i)
            action = stuff[1]
            actions.append(action)
            reward = stuff[2]
            state_next = stuff[3]
            if done:
                actualQ[i, action] = reward
            else:
                actualQ[i, action] = \
                reward + \
                self.gamma*np.max(self.model.predict(state_next.reshape(1,-1)))
        Q = self.model.predict(States)
        for i, j in zip(idx,actions):
            Q[i,j] = (1-self.alpha)*Q[i,j]
        Q += self.alpha*actualQ
        self.model.fit(States, Q, epochs=1, verbose=0)
```

```python
agent = Agent(gamma=gamma,
             epsilon=epsilon,
             d_epsilon=d_epsilon,
             alpha=alpha,
             n_states=n_states,
             n_actions=n_actions,
             history_len=history_len)

Rewards = deque(maxlen=100)
RewardsAll = []
n = 0
for episode in range(n_episodes):
    total_reward = 0
    state = env.reset()
    agent.epsilon_decay()
    for t in range(max_step):
#         env.render()
        action = agent.select_action(state)
        state_next, reward, done, _ = env.step(action)
        total_reward += reward
        agent.append_history(state, action, reward, state_next, done)
        agent.fit(train_set_size=64, steps=5)
        state = state_next
        if done: 
            Rewards.append(total_reward)
            RewardsAll.append(total_reward)
            print(
                'episode: {}, epsilon: {:.3f}, total rewards: {:.0f}, avg. rewards: {:.0f}'\
                .format(episode, agent.epsilon, total_reward, np.mean(Rewards))
            )
            if np.mean(Rewards)>200:
                n += 1
                agent.model.save(os.path.join(folder, 'episode{}.h5'.format(episode)))
            break
    if n==20:
        break
with open(os.path.join(folder, 'rewards.json'), 'w') as handle:
    json.dump({folder: RewardsAll}, handle)
```
